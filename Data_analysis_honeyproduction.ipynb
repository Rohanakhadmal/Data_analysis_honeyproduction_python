# 3. Display the percentage distribution of the data in each year using the pie chart.

```python
# So we have to distribute the data by 'year' and calculate the the percentage

yearly_dist = (df['year'].value_counts(normalize = True) * 100)
yearly_dist
```

```python
# Now we have to create a peichart

plt.figure(figsize=(5, 6))
plt.pie(yearly_dist, labels=yearly_dist.index, autopct='%1.1f%%', startangle=140)
plt.title('Percentage Distribution of Data by Year')
plt.axis('equal')
plt.show()
```

# 4. Plot and Understand the distribution of the variable "price per lb" using displot, and write your findings.

```python
# So now w have to plot the distribution of the variable "price per lb"

plt.figure(figsize=(8, 5))
sns.histplot(df['priceperlb'], kde=True, bins=20, color='grey')
plt.title('Distribution of Price per lb')
plt.xlabel('Price per lb')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

# 5. Plot and understand the relationship between the variables 'numcol' and 'prodval' through scatterplot, and write your findings.

```python
# So now we have to plot scatterploy to under the relationship between the variables 'numcol' and 'prodvalue'

plt.figure(figsize = (8, 5))
sns.scatterplot(x = df.numcol, y = df.prodvalue, color = "blue")
plt.title("Relatiobship between the variables 'numcol' and 'prodvalue'")
plt.xlabel("Number of colonies")
plt.ylabel("Number of Value")
plt.grid(True)
plt.show()
```

# 6. Plot and understand the relationship between categorical variable 'year' and a numerical variable 'prodvalue' through boxplot, and write your findings.

```python
# Plotting a boxplot to see the relationship between categorical variable 'year' and a numerical variable 'prodvalue'

plt.figure(figsize=(10, 5))
sns.boxplot(x= df.year, y=df.prodvalue)
plt.title('Relationship between Year and Production Value')
plt.xlabel('Year')
plt.ylabel('Production Value')
plt.grid(True)
plt.show()
```

# 7. Visualize and understand the relationship between the multiple pairs of variables throughout different years using pairplot and add your inferences. (use columns 'numcol', 'yield percol', 'total prod', 'prodvalue','year')


```python
# Now we have to select relevant column for our pairplot

col_of_interest = ['numcol', 'yieldpercol', 'totalprod', 'prodvalue', 'year']
data_subset = df[col_of_interest]
data_subset.head(10)
```

```python
# Now we have to plotting by pairplot

sns.pairplot(data_subset, hue='year', palette='Set1', diag_kind='kde')
plt.suptitle('Pairplot of Variables by Year', y=1.02)
plt.show()
```

# 8. Display the correlation values using a plot and add your inferences. (use columns 'numcol', 'yield percol', 'total prod', 'stocks', 'price per lb', 'prodvalue')

```python
col_interest = ['numcol', 'yieldpercol', 'totalprod', 'stocks', 'priceperlb', 'prodvalue']
data_subset = df[col_interest]
data_subset.head(10)
```

```python
# Now we have to compute the correlation matrix

corr_matrix = data_subset.corr()
corr_matrix
```

```python
# Now we have to create a heatmap to visualize the correlation matrix

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix of Selected Variables')
plt.show()
```
